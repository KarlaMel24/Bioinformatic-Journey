````python
import pandas as pd
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyRegressor
import numpy as np
````

## Pre-process the data

````python
df_x.shape
````

````python
y = df_y
y.shape
````

````python
# To remove the features who doesn't meet the threshold I want
from sklearn.feature_selection import VarianceThreshold
# Removes the columns with a variance lower than the threshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
# We have removed the columns with little variance (predictive power) and assigned
# The transformed dt to the variable x
x = selection.fit_transform(df_x)
````

````python
# Check the data dimension again
# We only do this with x, because is the variable with the descriptors column
x.shape
````

````python
# We split the data, so 80% is used to train the model and 20% to test it
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
````

## Testing ML models

We are going to check the r2 value in different models.

````python
# Define y builds the LazyRegressor
clf = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric=None)

# Train the model with the train data
models_train, predictions_train = clf.fit(x_train, x_train, y_train, y_train)

# Evaluates the model with the test data
models_test, predictions_test = clf.fit(x_train, x_test, y_train, y_test)
````

````python
# This is to check the performance of the training set
predictions_test
````

````python
predictions_test
````

Now, we are going to define our model.

````python
# Import model
from sklearn.ensemble import RandomForestRegressor

# Define model as RandomForestRegressor
RFR_model = RandomForestRegressor(n_estimators=100)
# Train model
RFR_model.fit(x_train, y_train)
#Calculate r2
r2 = RFR_model.score(x_test, y_test)
r2
````

````python
# Import model
from sklearn.svm import NuSVR

# Define the model as NuSRVR
model = NuSVR()
# Train the model
model.fit(x_train, y_train)
# Calculate r2
r2 = model.score(x_test, y_test)
r2
````

We can see that RandomForestRegressor doesn't have a very good r2, in the table is 0.39 and with our data is around 0.37. Therefore we will continue with NuSVR, which has the best r2 (.50 in both tables and our data).

````python
# Make predictions
y_pred = model.predict(x_test)
````

````python
#save model as a Pickle Object
import pickle
pickle.dump(model, open("E.coli_DR_model.pkl", "wb"))
````

The issue it's already solved in the first part of this project, but if want to delete a row that has an infinite value, this is how you can do it:

Firts, we see if we indeed have infinite numbers
```
import numpy as np
print(np.any(np.isinf(df_y)), np.any(np.isinf(df_x)))
```
Then we delate those rows
```
y_train = y_train.replace([np.inf, -np.inf], np.nan)
y_train = y_train.dropna()
```

After that we
```
inf_indices = np.isinf(y_train)
x_train = x_train[~inf_indices]
y_train = y_train[~inf_indices]
```
We confirm that we don't have infinite numbers anymore
```
print(np.any(np.isinf(y_train)), np.any(np.isinf(x_train)))
```
